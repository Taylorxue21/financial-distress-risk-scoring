```python
import pandas as pd
from sqlalchemy import create_engine
import os

# 1. Connect to the local PostgreSQL database (using environment variables to avoid exposing credentials)
db_user = os.getenv("DB_USER", "postgres")
db_password = os.getenv("DB_PASSWORD", "")
db_name = os.getenv("DB_NAME", "postgres")

db_url = f"postgresql://{db_user}:{db_password}@localhost:5432/{db_name}"
engine = create_engine(db_url)

# 2. Extract feature data (only valid data with alpha labels)
print("Extracting feature data from the database...")
query = """
    SELECT * FROM model_dataset 
    WHERE alpha IS NOT NULL
"""
df = pd.read_sql(query, engine)

# 3. Save the extracted data to a CSV file for offline use 
save_path = "../dataset.csv"
df.to_csv(save_path, index=False)

print(f"✅ Success! {len(df)} risk control data records have been exported as a static file.")
print(f"File saved at: {os.path.abspath(save_path)}")
```
