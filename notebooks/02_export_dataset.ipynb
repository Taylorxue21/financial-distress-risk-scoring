```python
import pandas as pd
from sqlalchemy import create_engine
import os

# 1. 连接你的本地 PostgreSQL 数据库
# (根据你之前的截图，用户名是 taylorxue，密码为空，直连 postgres 库)
engine = create_engine("postgresql://taylorxue:@localhost:5432/postgres")

# 2. 提取特征数据 (只提取有 alpha 标签的有效数据)
print("正在从数据库提取数据...")
query = """
    SELECT * FROM model_dataset 
    WHERE alpha IS NOT NULL
"""
df = pd.read_sql(query, engine)

# 3. 将数据保存为 dataset.csv 
# 🚨 极其重要：因为你当前在 notebooks/ 文件夹里，所以必须用 "../" 退回上一级
# 这样 dataset.csv 才会乖乖保存在你的项目根目录下（和 README.md 同级）
save_path = "../dataset.csv"
df.to_csv(save_path, index=False)

print(f"✅ 成功！已将 {len(df)} 条风控数据导出为静态文件。")
print(f"文件保存在: {os.path.abspath(save_path)}")
```
